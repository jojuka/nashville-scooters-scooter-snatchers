Python - Create Maps with Folium and Leaflet, Source:https://www.youtube.com/watch?v=QpBmO35pmVE, Code & Dogs

```python

import folium
#from folium.plugins import MarkerCluster
#from folium.plugins import FastMarkerCluster

m = folium.Map(location=[36.161030, -86.777191], zoom_start = 15) ##optional tiles = "Mapbox Bright", "Stamen Toner"
m
```

```python
folium.Marker(location=[36.166208, -86.771413], popup='Titans Stadium', tooltip='click for more info').add_to(m)
##optional - icon=folium.Icon(icon='cloud or envelope', color = 'red')
m
```

```python
folium.Marker(location=[36.107760, -86.812690], popup='Hillsboro Village', tooltip='click for more info', icon=folium.Icon(icon='cloud')).add_to(m)
m           
```

```python
folium.Marker(location=[36.102620, 86.827940], popup='The Gulch', tooltip='click for more info', icon=folium.Icon(icon='envelope', color = 'red')).add_to(m)
m
```

```python
locationScooter = [36.158981,-86.778382 ]
iconScooter = folium.features.CustomIcon('../data/MDHA_Promise_Zones/scooter.png', icon_size = (40, 40))
popupScooter = '<strong>scooter</strong>'
folium.Marker(location=locationScooter, tooltip='scooter', popup=popupScooter, icon = iconScooter).add_to(m)
m
```
