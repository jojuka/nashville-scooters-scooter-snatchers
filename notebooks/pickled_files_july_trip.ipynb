```python
import pandas as pd
import pickle
```

```python
%%time
july_trip = pd.read_csv('../data/july_trip.csv')
july_trip.head()
```

```python
july_trip.info()
```

```python
july_trip.companyName.unique()
```

```python
company_dict = {'Bird':0, 'Lyft': 1, 'Gotcha': 2, 'Lime': 3, 'SPIN': 4, 'JUMP': 5, 'Bolt Mobility': 6}
july_trip.companyName = july_trip.companyName.replace(company_dict)
july_trip.companyName = july_trip.companyName.astype(str).astype(int)
```

```python
july_trip.head()
```

```python
july_trip.pubTimeStamp = pd.to_datetime(july_trip.pubTimeStamp)
july_trip.startDate = pd.to_datetime(july_trip.startDate)
july_trip.endDate = pd.to_datetime(july_trip.endDate)
july_trip.info()
july_trip.head()
```

```python
july_trip = july_trip.drop(columns=['create_dt'])
july_trip.info()
```

```python
july_trip.to_pickle("../data/july_trip.pkl")
```
