### Create new dataframe and drop weekday rows

```python
# Copy may_df dataframe into new dataframe for weekends
may_wknd_df = may_clnd_df.copy()

# get names of indexes for weekends
index_names_wknd = may_wknd_df[may_wknd_df['weekday'] < 5].index

# drop weekday rows
may_wknd_df.drop(index_names_wknd, inplace = True)

#may_wknd_df.info()
may_wknd_df.head()
```

```python
#may_wknd_df.loc[((may_wknd_df['dff_prv_latitude'].abs() <= 0.002) & (may_wknd_df['dff_prv_longitude'].abs() <= 0.002)) | ((may_wknd_df['dff_prv_latitude'].isnull()) & (may_wknd_df['dff_prv_longitude'].isnull()))]
may_wknd_df.loc[((may_wknd_df['dff_prv_latitude'].abs() >= 0.002) | (may_wknd_df['dff_prv_longitude'].abs() >= 0.002)) | ((may_wknd_df['dff_nxt_latitude'].abs() >= 0.002) | (may_wknd_df['dff_nxt_longitude'].abs() >= 0.002)) | ((may_wknd_df['dff_prv_latitude'].isnull()) & (may_wknd_df['dff_prv_longitude'].isnull()))]
```

### Count of scooters per company

```python
may_wknd_co_cnt = may_wknd_df.groupby(['companyname'])['sumdid'].nunique().sort_values()
may_wknd_co_cnt
```

```python
may_wknd_co_bar = may_wknd_co_cnt.plot.bar('companyname', 'sumdid', rot=30, title='May 2019 Scooters').set_ylabel('count')
```
