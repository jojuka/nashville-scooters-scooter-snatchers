```python
import pandas as pd
import pickle
```

```python
%%time
may_trip = pd.read_csv('../data/may_trip.csv')
may_trip.head()
```

```python
may_trip.info()
```

```python
may_trip.companyName.unique()
```

```python
company_dict = {'Bird':0, 'Lyft': 1, 'Gotcha': 2, 'Lime': 3, 'SPIN': 4, 'JUMP': 5, 'Bolt Mobility': 6}
may_trip.companyName = may_trip.companyName.replace(company_dict)
may_trip.companyName = may_trip.companyName.astype(str).astype(int)
```

```python
may_trip.head()
```

```python
may_trip.pubTimeStamp = pd.to_datetime(may_trip.pubTimeStamp)
may_trip.startDate = pd.to_datetime(may_trip.startDate)
may_trip.endDate = pd.to_datetime(may_trip.endDate)
may_trip.info()
may_trip.head()
```

```python
may_trip = may_trip.drop(columns=['create_dt'])
may_trip.info()
```

```python
may_trip.to_pickle("../data/may_trip.pkl")
```
