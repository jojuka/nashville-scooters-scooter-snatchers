### Create new dataframe and drop weekday rows

```python
# Copy july_df dataframe into new dataframe for weekends
july_wknd_df = july_clnd_df.copy()

# get names of indexes for weekends
index_names_wknd = july_wknd_df[july_wknd_df['weekday'] < 5].index

# drop weekday rows
july_wknd_df.drop(index_names_wknd, inplace = True)

#july_wknd_df.info()
july_wknd_df.head()
```

```python
#july_wknd_df.loc[((july_wknd_df['dff_prv_latitude'].abs() <= 0.002) & (july_wknd_df['dff_prv_longitude'].abs() <= 0.002)) | ((july_wknd_df['dff_prv_latitude'].isnull()) & (july_wknd_df['dff_prv_longitude'].isnull()))]
july_wknd_df.loc[((july_wknd_df['dff_prv_latitude'].abs() >= 0.002) | (july_wknd_df['dff_prv_longitude'].abs() >= 0.002)) | ((july_wknd_df['dff_nxt_latitude'].abs() >= 0.002) | (july_wknd_df['dff_nxt_longitude'].abs() >= 0.002)) | ((july_wknd_df['dff_prv_latitude'].isnull()) & (july_wknd_df['dff_prv_longitude'].isnull()))]
```

### Count of scooters per company

```python
july_wknd_co_cnt = july_wknd_df.groupby(['companyname'])['sumdid'].nunique().sort_values()
july_wknd_co_cnt
```

```python
july_wknd_co_bar = july_wknd_co_cnt.plot.bar('companyname', 'sumdid', rot=30, title='july 2019 Scooters').set_ylabel('count')
```
