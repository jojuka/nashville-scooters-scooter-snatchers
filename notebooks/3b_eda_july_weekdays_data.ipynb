### Create new dataframe and drop weekends rows

```python
# Copy july_df dataframe into new dataframe for weekdays
july_wk_df = july_clnd_df.copy()

# get names of indexes for weekdays
index_names_wk = july_wk_df[july_wk_df['weekday'] >= 5].index

# drop weekend rows
july_wk_df.drop(index_names_wk, inplace = True)

#july_wknd_df.info()
july_wk_df.head()
```

### Find individual scooters per time per location

```python
#july_wk_df.loc[((july_wk_df['dff_prv_latitude'].abs() <= 0.002) & (july_wk_df['dff_prv_longitude'].abs() <= 0.002)) | ((july_wk_df['dff_prv_latitude'].isnull()) & (july_wk_df['dff_prv_longitude'].isnull()))]
july_wk_df.loc[((july_wk_df['dff_prv_latitude'].abs() >= 0.002) | (july_wk_df['dff_prv_longitude'].abs() >= 0.002)) | ((july_wk_df['dff_nxt_latitude'].abs() >= 0.002) | (july_wk_df['dff_nxt_longitude'].abs() >= 0.002)) | ((july_wk_df['dff_prv_latitude'].isnull()) & (july_wk_df['dff_prv_longitude'].isnull()))]
```

### Count of scooters per company

```python
july_wk_co_cnt = july_wk_df.groupby(['companyname'])['sumdid'].nunique().sort_values()
july_wk_co_cnt
```

```python
july_wk_co_bar = july_wk_co_cnt.plot.bar('companyname', 'sumdid', rot=30, title='july 2019 Scooters').set_ylabel('count')
```
