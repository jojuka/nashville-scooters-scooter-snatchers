### Create new dataframe and drop weekends rows

```python
# Copy june_df dataframe into new dataframe for weekdays
june_wk_df = june_clnd_df.copy()

# get names of indexes for weekdays
index_names_wk = june_wk_df[june_wk_df['weekday'] >= 5].index

# drop weekend rows
june_wk_df.drop(index_names_wk, inplace = True)

#june_wknd_df.info()
june_wk_df.head()
```

### Find individual scooters per time per location

```python
#june_wk_df.loc[((june_wk_df['dff_prv_latitude'].abs() <= 0.002) & (june_wk_df['dff_prv_longitude'].abs() <= 0.002)) | ((june_wk_df['dff_prv_latitude'].isnull()) & (june_wk_df['dff_prv_longitude'].isnull()))]
june_wk_df.loc[((june_wk_df['dff_prv_latitude'].abs() >= 0.002) | (june_wk_df['dff_prv_longitude'].abs() >= 0.002)) | ((june_wk_df['dff_nxt_latitude'].abs() >= 0.002) | (june_wk_df['dff_nxt_longitude'].abs() >= 0.002)) | ((june_wk_df['dff_prv_latitude'].isnull()) & (june_wk_df['dff_prv_longitude'].isnull()))]
```

### Count of scooters per company

```python
june_wk_co_cnt = june_wk_df.groupby(['companyname'])['sumdid'].nunique().sort_values()
june_wk_co_cnt
```

```python
june_wk_co_bar = june_wk_co_cnt.plot.bar('companyname', 'sumdid', rot=30, title='june 2019 Scooters').set_ylabel('count')
```
