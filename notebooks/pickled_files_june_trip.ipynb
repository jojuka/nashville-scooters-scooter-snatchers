```python
import pandas as pd
import pickle
```

```python
%%time
june_trip = pd.read_csv('../data/june_trip.csv')
june_trip.head()
```

```python
june_trip.info()
```

```python
june_trip.companyName.unique()
```

```python
company_dict = {'Bird':0, 'Lyft': 1, 'Gotcha': 2, 'Lime': 3, 'SPIN': 4, 'JUMP': 5, 'Bolt Mobility': 6}
june_trip.companyName = june_trip.companyName.replace(company_dict)
june_trip.companyName = june_trip.companyName.astype(str).astype(int)
```

```python
june_trip.head()
```

```python
june_trip.pubTimeStamp = pd.to_datetime(june_trip.pubTimeStamp)
june_trip.startDate = pd.to_datetime(june_trip.startDate)
june_trip.endDate = pd.to_datetime(june_trip.endDate)
june_trip.info()
june_trip.head()
```

```python
june_trip = june_trip.drop(columns=['create_dt'])
june_trip.info()
```

```python
june_trip.to_pickle("../data/june_trip.pkl")
```
