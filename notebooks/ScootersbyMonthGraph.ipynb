```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/month.csv')
```

```python
print (df)
```

```python
df = df.sort_values('Company')
```

```python
df.pivot(index='Company', columns='Month', values = 'Value').plot(kind='bar',figsize=(15,8))
plt.xlabel('Company')
plt.ylabel('Number of Scooters')
plt.title('Scooters by Month, May to July, 2019')
plt.legend(loc='upper right')
plt.show()
```
