### Create new dataframe and drop weekday rows

```python
# Copy june_df dataframe into new dataframe for weekends
june_wknd_df = june_clnd_df.copy()

# get names of indexes for weekends
index_names_wknd = june_wknd_df[june_wknd_df['weekday'] < 5].index

# drop weekday rows
june_wknd_df.drop(index_names_wknd, inplace = True)

#june_wknd_df.info()
june_wknd_df.head()
```

```python
#june_wknd_df.loc[((june_wknd_df['dff_prv_latitude'].abs() <= 0.002) & (june_wknd_df['dff_prv_longitude'].abs() <= 0.002)) | ((june_wknd_df['dff_prv_latitude'].isnull()) & (june_wknd_df['dff_prv_longitude'].isnull()))]
june_wknd_df.loc[((june_wknd_df['dff_prv_latitude'].abs() >= 0.002) | (june_wknd_df['dff_prv_longitude'].abs() >= 0.002)) | ((june_wknd_df['dff_nxt_latitude'].abs() >= 0.002) | (june_wknd_df['dff_nxt_longitude'].abs() >= 0.002)) | ((june_wknd_df['dff_prv_latitude'].isnull()) & (june_wknd_df['dff_prv_longitude'].isnull()))]
```

### Count of scooters per company

```python
june_wknd_co_cnt = june_wknd_df.groupby(['companyname'])['sumdid'].nunique().sort_values()
june_wknd_co_cnt
```

```python
june_wknd_co_bar = june_wknd_co_cnt.plot.bar('companyname', 'sumdid', rot=30, title='june 2019 Scooters').set_ylabel('count')
```
