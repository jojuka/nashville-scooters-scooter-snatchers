### Create new dataframe and drop weekends rows

```python
# Copy may_df dataframe into new dataframe for weekdays
may_wk_df = may_clnd_df.copy()

# get names of indexes for weekdays
index_names_wk = may_wk_df[may_wk_df['weekday'] >= 5].index

# drop weekend rows
may_wk_df.drop(index_names_wk, inplace = True)

#may_wknd_df.info()
may_wk_df.head()
```

### Find individual scooters per time per location

```python
#may_wk_df.loc[((may_wk_df['dff_prv_latitude'].abs() <= 0.002) & (may_wk_df['dff_prv_longitude'].abs() <= 0.002)) | ((may_wk_df['dff_prv_latitude'].isnull()) & (may_wk_df['dff_prv_longitude'].isnull()))]
may_wk_df.loc[((may_wk_df['dff_prv_latitude'].abs() >= 0.002) | (may_wk_df['dff_prv_longitude'].abs() >= 0.002)) | ((may_wk_df['dff_nxt_latitude'].abs() >= 0.002) | (may_wk_df['dff_nxt_longitude'].abs() >= 0.002)) | ((may_wk_df['dff_prv_latitude'].isnull()) & (may_wk_df['dff_prv_longitude'].isnull()))]
```

### Count of scooters per company

```python
may_wk_co_cnt = may_wk_df.groupby(['companyname'])['sumdid'].nunique().sort_values()
may_wk_co_cnt
```

```python
may_wk_co_bar = may_wk_co_cnt.plot.bar('companyname', 'sumdid', rot=30, title='May 2019 Scooters').set_ylabel('count')
```
